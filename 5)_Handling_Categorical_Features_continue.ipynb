**Ordinal Number Encoding**

An ordinal encoding involves mapping each unique label to an integer value.

This type of encoding is really only appropriate if there is a known relationship between the categories.

```python
import datetime
```

```python
today_date= datetime.datetime.today()
```

```python
today_date
```

```python
#Timedelta Represent the difference between two datetime objects.

today_date-datetime.timedelta(1)
#today-1 day difference
```

```python
today_date-datetime.timedelta(2)
#today-2 day difference
```

```python
days=[]
for i in range(0,15):
  days.append(today_date-datetime.timedelta(i))
```

```python
days
```

```python
import pandas as pd

data= pd.DataFrame(data=days, columns=['Day'])
data
```

```python
data['WeekDay']= data['Day'].dt.day_name()
data['Date']= data['Day'].dt.date
data['Year']= data['Day'].dt.year
data['Month']= data['Day'].dt.month
data
```

```python
dictionary={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
dictionary
```

```python
data['WeekDay_Ordinal']= data['WeekDay'].map(dictionary)
data
```

**Count Or Frequency Encoding**

Mapping the categories with its frequency in a given feature columns

```python
df= pd.read_csv('/content/mercedes.csv', usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df.head()
```

```python
for i in df.columns:
  print('Feature: ', i, ', Categories: ', len(df[i].unique()))
```

```python
df['X0'].value_counts()
```

```python
X0_map= df['X0'].value_counts().to_dict()
X0_map
```

```python
df['X0_freq_map']= df['X0'].map(X0_map)
```

```python
df
```

**Advantages**
1. Easy To Use
2. Not increasing feature space 

**Disadvantages**
1. It will provide same weight if the frequencies are same

**Target Guided Ordinal Encoding**
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv('/content/titanic_train.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df.info()
```

```python
df['Cabin'].unique()
```

```python
len(df['Cabin'].unique())
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'].astype('str')
```

```python
#Grabbing the first letter of categories:
df['Cabin'].astype('str').str[0]
```

```python
df['First_Letter']= df['Cabin'].astype('str').str[0]
```

```python
df
```

```python
df['First_Letter'].unique()
```

```python
df.groupby(by=['First_Letter'])['Survived'].mean().sort_values()
#Here we are sorting the value according to its means to extract thge importace of the category with respect to target(survived).
```

```python
ordinal_labels= df.groupby(by=['First_Letter'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels_imp= {k:i for i,k in enumerate(iterable=ordinal_labels, start=0)}
ordinal_labels_imp
```

```python
df['Ordinal_with_Importance']= df['First_Letter'].map(ordinal_labels_imp)
```

```python
df
```

**Mean Encoding**

```python
mean_ordinal= df.groupby(by=['First_Letter'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['Mean_Ordinal']= df['First_Letter'].map(mean_ordinal)
```

```python
df
```

**Probability Ratio Encoding**

1. Probability of Survived based on Cabin--- Categorical Feature
2. Probability of Not Survived---1-pr(Survived)
3. pr(Survived)/pr(Not Survived)
4. Dictonary to map cabin with probability
5. replace with the categorical feature

```python
df=pd.read_csv('titanic_train.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['First_Letter']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['First_Letter'].unique()
```

```python
prob_df=df.groupby(['First_Letter'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
probability_encoded=prob_df['Probability_ratio'].to_dict()
probability_encoded
```

```python
df['Cabin_with_Prob']=df['First_Letter'].map(probability_encoded)
df.head()
```
