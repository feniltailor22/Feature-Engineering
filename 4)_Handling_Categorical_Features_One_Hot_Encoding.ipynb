```python
#To apply One Hot Encoding to Many Categories in a Single feature will be computationaly expensive.
```

```python
#To solve this, we will only consider Top 10 frequent categories in each feature column and will drop remaining categories.
```

```python
#Obtaining Top 10 Categories for 'X1'

df['X1'].value_counts().head(10)
#The resulting object will be in descending order so that the first element is the most frequently-occurring element.
```

```python
#grabbing the index of Top 10 X1 Categories
list_X1= list(df['X1'].value_counts().head(10).index)
list_X1
```

```python
for i in list_X1:
  df[i]= np.where(df['X1']==i, 1, 0)
```

```python
df[list_X1]
```

```python
list_X1.append('X1')
```
